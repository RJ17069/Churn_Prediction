### Data Cleaning

```python
telco_data = telco_base_data.copy()
```

```python
telco_data.TotalCharges = pd.to_numeric(telco_data.TotalCharges, errors = 'coerce')
telco_data.isnull().sum()
```

```python
missing = pd.DataFrame((telco_data.isnull().sum())*100/telco_data.shape[0]).reset_index()
plt.figure(figsize=(16,5))
ax = sns.pointplot(x='index',y=0,data=missing)
plt.xticks(rotation =90,fontsize =7)
plt.title("Percentage of Missing values")
plt.ylabel("PERCENTAGE")
plt.show()
```

```python
telco_data.loc[telco_data['TotalCharges'].isnull() == True]
```

```python
telco_data.dropna(how = "any", inplace = True)
#telco_data.fillna(0)
```

```python
telco_data.isnull().sum()
```

```python
labels = ["{0} - {1}".format(i, i+11)for i in range(1, 72, 12)]
telco_data['tenure_group'] = pd.cut(telco_data.tenure, range(1, 80, 12), right = False, labels = labels)
```

```python
telco_data['tenure_group'].value_counts()
```

```python
telco_data.drop(columns=['customerID','tenure'], axis=1, inplace = True)
telco_data.head()
```

```python
for i, prediction in enumerate(telco_data.drop(columns=['Churn', 'TotalCharges','MonthlyCharges'])):
    plt.figure(i)
    sns.countplot(data = telco_data, x = prediction, hue = 'Churn')
```

```python
telco_data['Churn'] = np.where(telco_data.Churn == 'Yes', 1, 0)
telco_data.head()
```

```python
telco_data_dummy = pd.get_dummies(telco_data)
telco_data_dummy.head()
```

```python
sns.lmplot(data = telco_data_dummy, y = 'TotalCharges', x = 'MonthlyCharges')
```

```python
Mth1 = sns.kdeplot(telco_data_dummy.MonthlyCharges[(telco_data_dummy['Churn'] == 0)],
                  color = 'Red', fill = True)
Mth = sns.kdeplot(telco_data_dummy.MonthlyCharges[(telco_data_dummy['Churn'] == 1)],
                  ax = Mth1, color = 'Blue', fill = True)
Mth.legend(["No Churn","Churn"],loc='upper right')
Mth.set_xlabel('Monthly Charges')
Mth.set_ylabel('Density')
Mth.set_title('Monthly charges by churn')
```

### Bivariate Analysis

```python
new_df0 = telco_data.loc[telco_data['Churn'] == 0]
new_df1 = telco_data.loc[telco_data["Churn"] == 1]
```

```python
def uniplot(df,col,title,hue =None):
    
    sns.set_style('whitegrid')
    sns.set_context('talk')
    plt.rcParams["axes.labelsize"] = 20
    plt.rcParams['axes.titlesize'] = 22
    plt.rcParams['axes.titlepad'] = 30
    
    
    temp = pd.Series(data = hue)
    fig, ax = plt.subplots()
    width = len(df[col].unique()) + 7 + 4*len(temp.unique())
    fig.set_size_inches(width , 8)
    plt.xticks(rotation=45)
    plt.yscale('log')
    plt.title(title)
    ax = sns.countplot(data = df, x= col, order=df[col].value_counts().index,hue = hue,palette='bright') 
        
    plt.show()
```

```python
uniplot(new_df1,col='Partner',title='Distribution of Gender for Churned Customers',hue='gender')
```

```python
uniplot(new_df0,col='Partner',title='Distribution of Gender for Non Churned Customers',hue='gender')
```
